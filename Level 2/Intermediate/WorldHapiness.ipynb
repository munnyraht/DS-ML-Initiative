```python
def rename_columns(df, column):
    for column_name in df.columns:
        if column.lower() in column_name.lower() or column_name.lower() in column.lower():
            df.rename({column_name : column}, axis = 1, inplace= True)
            return
```

```python
print(df_2015.columns)
print(df_2016.columns)
print(df_2017.columns)
print(df_2018.columns)
print(df_2019.columns)
```

```python
df_2015['Year'] = '2015'
df_2016['Year'] = '2016'
df_2017['Year'] = '2017'
df_2018['Year'] = '2017'
df_2019['Year'] = '2019'
```

```python
Merged_df['Trust'].fillna((Merged_df['Trust'].mean()), inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,confusion_matrix
```

```python
X = Merged_df.drop(['Happiness Score','Country','Year'], axis=1)
y = Merged_df['Happiness Score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
Linear_Model = LinearRegression()
```

```python
Linear_Model.fit(X_train,y_train)
```

```python
y_pred = Linear_Model.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

Finally, A model that predicts a country happiness score according to some features has been built , it is about 73% accurate.
